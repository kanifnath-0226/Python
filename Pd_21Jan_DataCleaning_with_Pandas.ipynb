#### Pandas
There are 2 data structures

#### 1) Series
a) A series consists of Single Column data.<br>
b) A series will always bear index. There can be multiple indexes as well.


#### 2) DataFrame
a) A DataFrame consists of data organized into Rows and columns. A DataFrame consists of multiple Series<br>
b) A DataFrame will always bear index. There can be multiple indexes as well.

#### Note
A series will always contain a single column. However a dataframe can consist of single column as well. But Series cant contain multiple columns.

#### Creating a dataframe using Dictionary

The Key used for dictionary will be used as column names. Values will be used as data of the dataframe. Values should be initialized with list of same length.

```python
# np.nan = Null value

data = {'SID': [i for i in range(1,13)],
       'Name': ['Utsav','Paritosh','Aditya','Aryan','Ankita','Jyotsana',
              'Manan','Shivangi','Garvit','Harsh','Ayushi','Karan'],
       'Age': [23,24,np.nan,26,28,np.nan,30,22,24,25,20,np.nan],
       'City': ['Mumbai','Delhi','Noida','Delhi',np.nan,'Noida','Mumbai',
              'Delhi','Noida','Pune','Delhi','Mumbai'],
       'Marks': [87,np.nan,78,67,81,np.nan,93,np.nan,70,65,88,59],
       'Subject': [np.nan,'Stats','SQL',np.nan,'ML','Python','ML',
                 'Python','SQL',np.nan,'Python','Python']
       }

df = pd.DataFrame(data)
df
```

#### Extract Year, Quarter, Month, Month_Name, Day, Day_Name from Start_Date

```python
# dt = datetime

df['Year'] = df['Start_Date'].dt.year
df['Qtr'] = df['Start_Date'].dt.quarter
df['Month'] = df['Start_Date'].dt.month
df['Month_Name'] = df['Start_Date'].dt.month_name()
df['Day'] = df['Start_Date'].dt.day
df['Day_Name'] = df['Start_Date'].dt.day_name()
df['WeekDay'] = df['Start_Date'].dt.weekday # Mon-0, Sun- 6
df.head(7)
```

#### 2) Handling Null Values

#### Null Value Treatment
1) Remove columns conatining null values. <br>
2) Remove rows containing null values. <br>
3) Fill Null values(Imputation) with appropriate data. <br>

#### Imputation (Filling Null Values)
1) Continuous Data - Fill missing values with Mean or Median<br>
2) Categorical Data - Fill missing values with Mode or 'Unknown' <br>
3) Time Series Data - Fill missing values with bfill (backward fill) or ffill (forward fill).<br>

```python
df2 = df.iloc[:,:7]
df2.head()
```

```python
df2.dtypes
```

#### Detect null values

```python
df2.info()
```

```python
df2.isnull().sum() 
# Returns a Series conatining column wise sum of null values
```

```python
df2['Age'] = df2['Age'].fillna(df2['Age'].mean())

# OR

df2['Marks'].fillna(df2['Marks'].median(),inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2.shape
```

```python
df2['City'].value_counts(dropna=False)
```

```python
df2['Subject'].value_counts(dropna=False)
```

```python
df2['City'].fillna(df2['City'].mode()[0],inplace=True)
df2['Subject'].fillna('Unknown',inplace=True)  # Not Available - NA, None
```

```python
df2.isnull().sum()
```

#### Handling Duplicate Rows

```python
### 1) Detect duplicates
### 2) Drop Duplicates
```
