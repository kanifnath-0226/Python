Q1) Create a simple Bar plot of name and total_marks.

```python
#plt.figure(figsize=(20,8))
plt.bar(ass['Name'],ass['Total_marks'],label='marks',color='green')
plt.title("Name vs Total marks",size=25,color='red')
plt.legend()
plt.xlabel('Names',size=20)
plt.ylabel('Total marks',size=20)
plt.xticks(size=15,rotation=45)
plt.yticks(size=15, )
plt.grid()
plt.show()
```

4.Create a bar plot of 'Name' vs 'Statistics' and apply the Following customization to the plot.

Show legend with label as 'Statistics' <br>

```python
s=list(range(8)) 

plt.figure(figsize=(13,4))
plt.bar(ass['Name'],ass['Statistics'],color='cyan',edgecolor='black',label='name vs Statistics')
plt.title("Names vs 'Statistics'",size=15,color='red')
plt.legend()
plt.ylim(0,120)
plt.xlabel('Names')
plt.ylabel('Statistics')
plt.xticks(rotation=0)
#plt.savefig('Name_vs_SQL.png')
 
for  idx ,val in enumerate(ass['Statistics']): 
    plt.annotate(xy=[s[idx]-0.2,int(val)+2],text=int(val))

plt.show()
```
