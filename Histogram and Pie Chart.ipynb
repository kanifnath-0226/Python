### Histogram 
1) It is used to represent frequency distribution.<br>
2) One axis represents the varaibles in form of bars, the other represents the freqency of each bar

<b>plt.hist(x,color,bins,edgecolor)</b>
    
x - data(column) for which we need to compute the frequency<br>
color - Used to sepecify the color of the bars of the histogram<br>
bins - Buckets fo which the frequency will be computed<br>
edgecolor - Used to sepecify the edgecolor of the bars of histogram<br>

```python
x = [1,2,3,2,4,2,3,1,5,6,4,4,3,5,7,3,6,7,2,3,4,7]
print(len(x))
```

```python
from collections import Counter
c1 = Counter(x)
print(c1)
```

```python
plt.hist(x,color='orange')
plt.show()
```

```python
plt.hist(x,bins=[1,2,3,4,5,6,7],color='orange',edgecolor='black')
print(c1)
plt.show()
# 1st bin : [1,2) => Including count of 1,excluding count of 2 => 2
# 2nd bin : [2,3) => Including count of 2,excluding count of 3 => 4
# 3rd bin : [3,4) => Including count of 3,excluding count of 4 => 5
# 4th bin : [4,5) => Including count of 4,excluding count of 5 => 4
# 5th bin : [5,6) => Including count of 5,excluding count of 6 => 2
# 6th bin : [6,7] => Including count of 6,including count of 7 => 2 + 3 = 5
```

```python
plt.hist(x,bins=[1,2,3,4,5,6,7,8],color='orange',edgecolor='black')
print(c1)
plt.show()
# 1st bin : [1,2) => Including count of 1,excluding count of 2 => 2
# 2nd bin : [2,3) => Including count of 2,excluding count of 3 => 4
# 3rd bin : [3,4) => Including count of 3,excluding count of 4 => 5
# 4th bin : [4,5) => Including count of 4,excluding count of 5 => 4
# 5th bin : [5,6) => Including count of 5,excluding count of 6 => 2
# 6th bin : [6,7) => Including count of 6,excluding count of 7 => 2
# 7th bin : [7,8] => Including count of 7,including count of 8 => 3
```

### Pie Chart
1) It is used to represent precentage distribution.
2) To plot pie chart, we need a continous column and its corresponding labels (the categories that this continous column highlights)

<b>plt.pie(x,labels,color,explode,autopct,shadow)</b>

x - data(continuous column) for which percentage values will be computed<br>
labels - Categorical description for the continuous column x. It will be represented as legend<br>
color - Used to specify the color of the different pies<br>
autopct - Used to display the precentage values in a particular format<br>
shadow - Used to set the shadow on or off<br>
counterclock - Used to display the values anitclockwise or clockwise<br>

#### Q) Represent the percentage distribution of mean of profit  based on title_year on a Pie Chart. 
Where Proift = Gross - Budget

```python
d1['Profit'] = d1['Gross'] - d1['budget']
d1.head(3)
```

```python
q3 = d1.groupby('title_year')['Profit'].mean().reset_index()
q3
```

```python
sum_profit = q3['Profit'].sum()
print(sum_profit)

per_2015 = 107.508278/sum_profit
print(per_2015*100)
```

```python
plt.pie(x = q3['Profit'],labels=q3['title_year'],autopct='%.2f%%')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(x = q3['Profit'],labels=q3['title_year'],
        colors=['red','blue','yellow','tan','brown','green','cyan'],autopct='%.3f%%',
       explode=(0,0,0,0,0,0,0.1),shadow=True)
plt.title('Year wise Mean of Profit in Million dollars',fontsize=15,color='blue')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(x = q3['Profit'],labels=q3['title_year'],
        colors=['red','blue','yellow','tan','brown','green','cyan'],autopct='%.3f%%',
       explode=(0,0,0,0,0,0,0.1),shadow=True,counterclock=False)
plt.title('Year wise Mean of Profit in Million dollars',fontsize=15,color='blue')
plt.show()
```
