## Line Chart
1)it is used to represent the variation of a continuous data over time<br>
2)however,it can be used with categorical columns as well.

1.Plot count of genre2 on a line chart where count>1

```python
q1 = df['genre_2'].value_counts()
q1 = q1[q1>1]
q1
```

```python
plt.figure(figsize=(15,5))
plt.plot(q1.index,q1.values,color='blue',marker='*',markersize = 14 ,label = 'count',lw =3 ,linestyle = ':')
plt.xlabel('Genre2',size = 14)
plt.ylabel('Count',size = 14)
plt.legend()
plt.title("Genre2 vs Count",fontsize = 15,color = 'red')
plt.grid()
 
plt.show()
```

2.plot year wise Profit on a line chart where Profit is Gross- budget

```python
d1['Gross'] = d1['Gross']/1000000
d1['budget'] = d1['budget']/ 1000000
d1.head()
```

```python
d1['profit'] = d1['Gross']-d1['budget']
d1.head()
```

```python
q2 = d1.groupby('title_year')['profit'].mean().reset_index()
q2 = q2.sort_values(by = 'title_year',ascending=True) 
q2
```

```python
plt.figure(figsize=(10,4))
plt.plot(q2['title_year'],q2['profit'],color='brown',marker='*',label = 'profit')
plt.xlabel('title_year')
plt.ylabel('Profit')
plt.grid()
plt.legend()
plt.title("Year wise avg profit(in M)",size = 15,color = 'Blue')
plt.show()
```

```python
q3 = d1.groupby('title_year')[['budget','profit']].mean().reset_index()
q3 = q3.sort_values(by = 'title_year',ascending=True) 
q3.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(q3['title_year'],q3['budget'],color='brown',marker='*',label = 'budget')
plt.plot(q3['title_year'],q3['profit'],color='green',marker='*',label = 'profit')
plt.xlabel('Title_year',size = 15,color = 'black')
plt.ylabel('Profit',size=15,color = 'black')
plt.grid()
plt.legend()
plt.title("Year wise mean Budget and profit(in M)",size = 15,color = 'Blue')
plt.show()
```

### Histogram
1) it is used to represent distribution

```python
from collections import Counter
```

```python
x = [1,2,3,4,1,2,3,4,5,6,7,8,7,6,5,4,3,2,1,2,3,5,4,5,6,7,8,7,6,5,4,3,2,1,3,4,5,6,7,3]
print(len(x))
```

```python
c1 = Counter(x)
print(c1)
```

```python
plt.figure(figsize=(15,5))
plt.hist(x,color='blue',bins=[1,2,3,4,5,6,7,8],edgecolor='black',label = 'count')

print(c1)
plt.legend()
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()
```

## pie chart

1) it is used to represent percenatage distribution.<br>
2) to plot the piechart we need a continuous column and its corresponding labels 

Q.represent the percentage distribution of year wise mean of profit on a pie chart .<br>where Profit = Gross -budget

```python
d2['profit']=d2['Gross']-d2['budget']
d2.head()
```

```python
q3 = d2.groupby('title_year')['profit'].mean().reset_index()
q3.head()
```

```python
sum_pro=q3['profit'].sum()
sum_pro
```

```python
plt.pie(x=q3['profit'],labels=q3['title_year'],autopct='%.2f%%')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(x=q3['profit'],labels=q3['title_year'],
        colors=['green','cyan','blue','yellow','pink','brown','red'],
        autopct='%.2f%%',explode=(0,0,0,0,0,0,0.1),shadow=True,counterclock=False)
plt.title('Year wise Mean of porfit in million dollars ',size=15,color='blue')
plt.show()
```
