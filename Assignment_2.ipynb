1.Create  a simple histogram for total marks .

```python
plt.hist(s1['Total_marks'],color='green',label='marks',edgecolor='black')
plt.title('Total_marks',size=15,color='black')
plt.legend()
plt.show()
```

2)

```python
plt.figure(figsize=(9,4))
plt.hist(s1['Total_marks'],color='teal',bins=10,label='marks',edgecolor='black')
plt.title('Total_marks',size=15,color='purple')
plt.xlabel('Total Marks')
plt.legend()
plt.show()
```

3)

```python
plt.figure(figsize=(5,5))
plt.pie(labels=s1['Name'],x=s1['Python'],autopct='%.2f%%')
plt.title("Name wise Python Marks ",size= 15 , color='teal')
plt.show()
```

4)

```python
plt.figure(figsize=(5,5))
plt.pie(labels=s1['Name'],x=s1['Total_marks'],autopct='%.2f%%',
        colors=['red','blue','yellow','green','cyan','deeppink','dodgerblue','orange'],
        explode=(0,0,0,0,0,0.2,0,0.1),shadow=True,counterclock=False)
plt.title("Name wise Python Marks ",size= 15 , color='teal')
plt.legend(loc='right',bbox_to_anchor=(1, 1, 0.6, -0.6) )

plt.show()
```

5)

```python
plt.boxplot(s1['Total_marks'])
plt.grid()
plt.show()
```

6)

```python
plt.figure(figsize=(10,4))
 
plt.bar(s1['Name'],s1['Python'],color='teal',label='Python',edgecolor='black',alpha=0.3)
plt.bar(s1['Name'],s1['SQL'],color='orange',label='SQL',edgecolor='black',alpha=0.5)
plt.title('Name wise Python and SQL marks',size=15,color='blue')
plt.legend( )
plt.xlabel("Names")
plt.ylabel('Marks')
plt.grid()
plt.ylim(0,175)
```

```python
r2 = s1[['Name','Python','SQL' ]]
r2.head()
r2.set_index('Name',inplace=True )
```

```python
r2.plot(kind='bar',stacked=True)
plt.grid()
plt.title('Name vs Pyhton and SQL Marks',size=15,color='blue')
plt.xticks(rotation=0)
plt.ylim(0,250)
plt.xlabel("Names")
plt.ylabel('marks')
plt.show()
```
