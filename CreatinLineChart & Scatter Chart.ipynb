### Line Chart
1) It is used to represent the variation of a continuous data over time<br>
2) However, it can be used with categroical columns as well<br>

<b>plt.plot(x,y,color,lw,label,linestyle,marker)</b><br>

x - data(column) to be plotted on the x-axis<br>
y - data(column) to be plotted on the y-axis<br>
color - color of the line<br>
lw - linewdith<br>
linestyle - types of line  <br>
marker - highlights the connection between x and y<br>
markersize - Determines the szie of the marker

Types of Markers: * , o, 1, 2, 3, d, <, >, ^ etc<br>
Types of Linestyle:    --(dashed), -.(dash dotted), :(dotted)

#### Q) Plot count of genre2 on a line chart where count >1

```python
q1 = df['genre_2'].value_counts() 
q1 = q1[q1>1]
print(q1)
```

```python
print(q1.index)
print(q1.values)
```

```python
plt.figure(figsize=(9,4))
plt.plot(q1.index,q1.values,color='maroon',marker='o',
         label='count',markersize=12,lw=3,linestyle='--')
plt.xlabel('Genre2')
plt.ylabel('Count')
plt.legend()
plt.title('Genre2 vs Count',fontsize=15,color='Blue')
plt.grid()
plt.show()
```

#### Q) Plot year wise mean of Profit on a line chart where Profit = Gross - budget

Converting Budget and Gross from dollars to Millon dollars

```python
d1["Gross"] = d1["Gross"]/1000000
d1["budget"] = d1["budget"]/1000000
d1.head()
```

```python
d1['Profit'] = d1['Gross'] - d1['budget']
d1.head()
```

```python
q2 = d1.groupby('title_year')['Profit'].mean().reset_index()
q2 = q2.sort_values(by='title_year',ascending=True)
q2
```

```python
plt.plot(q2['title_year'],q2['Profit'],color='brown',marker='*',markersize=12)
plt.title('Year wise avg Profit(in M)',fontsize=13,color='Blue')
plt.xlabel('Title_Year')
plt.ylabel('Profit')
plt.grid()
plt.show()
```

```python
q3 = d1.groupby('title_year')['budget','Profit'].mean().reset_index()
q3 = q3.sort_values(by='title_year',ascending=True)
q3
```

```python
plt.plot(q3['title_year'],q3['budget'],color='green',marker='*',label='Budget',markersize=8)
plt.plot(q3['title_year'],q3['Profit'],color='brown',marker='o',label='Profit',markersize=10)
plt.title('Year wise mean Budget and Profit(in M)',fontsize=13,color='Blue')
plt.xlabel('Title_Year')
# plt.ylabel('Profit')
plt.legend()
plt.grid()
plt.show()
```

### Scatter Chart
1) It is used to represent the distribution of the data<br>
2) It gives us the idea of covariance amongst the continuous fields plotted<br>
3) It is only meant for continuous data(columns)<br>


<b> plt.scatter(x,y,color,label,marker,s)</b><br>
    
x - data(column) to be plotted on the x-axis<br>
y - data(column) to be plotted on the y-axis<br>
color - color of the line<br>
marker - highlights the connection between x and y<br>
s - Determines the szie of the marker

Types of Markers: * , o, 1, 2, 3, d, <, >, ^ etc<br>

#### Q) Plot a scatter chart between Proift and Budget

```python
plt.scatter(d1['Profit'],d1['budget'],marker='^',s=50,color='orange')
plt.xlabel('Profit')
plt.ylabel('Budget')
plt.title('Profit vs Budget',fontsize=15,color='Maroon')
plt.grid()
plt.show()
```

```python
plt.scatter(d1['Profit'],d1['budget'],marker='^',s=50,color='orange',label='Profit vs budget')
plt.scatter(d1['Profit'],d1['Gross'],marker='o',s=45,color='lightblue',label='Profit vs Gross')
plt.xlabel('Profit')
plt.title('Profit vs Budget and Gross',fontsize=15,color='black')
plt.grid()
plt.legend()
plt.show()
```
